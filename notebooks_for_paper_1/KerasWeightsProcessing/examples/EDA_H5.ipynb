```python
import h5py
import pprint as pp
```

If there are no weight files, first run `python test_network.py` to generate one. Then add the filename here to examin it

```python
file_name = 'sequential_5_relu_0.0_False.h5'
```

```python
with h5py.File(file_name, 'r') as weights_file:
    print(weights_file.keys())
    print([i for i in weights_file.attrs])
    
    model_weights = weights_file['model_weights']
    
    model_config = weights_file.attrs['model_config'].decode('utf-8') # Decode using the utf-8 encoding
    model_config = model_config.replace('true','True')
    model_config = model_config.replace('false','False')

    model_config = model_config.replace('null','None')
    model_config = eval(model_config)
    
#     print(sorted(model_weights['batch_normalization_1']['batch_normalization_1'].keys()))
    training_config = eval(
        weights_file.attrs['training_config'].decode('utf-8').replace('null','None').replace('false','False').replace('true','True')
    )
    
    print('Learning rate:', training_config['optimizer_config']['config']['learning_rate'])
    pp.pprint(model_config)
```
